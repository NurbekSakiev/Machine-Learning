```python
import numpy as np
from numpy import convolve
import matplotlib.pyplot as plt
 
def movAverage (array, size):
    weights = np.repeat(1.0, size)/size
    simple_moving = np.convolve(array, weights, 'valid')
    return simple_moving

dataset= [1,9,10,12,14,19,20,23,25,28]
sample = [1,2,3,5,7,10,13,17,20,21]
 
s_movAver = movAverage(sample,3)
print (s_movAver)
```

```python
plt.plot(dataset,sample)
plt.show()
```

```python
plt.plot(dataset[len(dataset)-len(s_movAver):],s_movAver)
plt.show()
```

```python
plt.plot(dataset[len(dataset)-len(s_movAver):],s_movAver)
plt.plot(dataset,sample)
plt.show()
```
